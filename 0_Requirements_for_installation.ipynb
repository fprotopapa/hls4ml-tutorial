### Python

Finally, we should be using Python3 and the venv package, these can be installed by running following commands inside the console.
```
sudo apt install python3 python3-venv
```
The version used is 3.6.9.

````python
%%bash
python3 -V
````

### Git

To use github easily , git can be installed with
```
sudo apt install git
```
Clone this repository with following command 

````python
git clone https://github.com/fprotopapa/hls4ml-tutorial.git
````

### Virtual enviroment and directory structure

```
These steps would be normally done inside the console window.
```
Now let's create a project folder hls4ml and change the working directory.

````python
%mkdir hls4ml
%cd hls4ml
````

To keep the system-wide Python installation clean, let's create a virtual enviroment, by running:

````python
%%bash
python3 -m venv venv_hls4ml
````

### Jupter notebook

To follow along with jupyter notebook, we have to install it. 
This also requires pip, which is the python package-management system.
The code below will activate the virtual enviroment, install jupyter and add the venv to it.

````python
%%bash
source venv_hls4ml/bin/activate
pip install --upgrade pip
pip install jupyterlab
ipython kernel install --user --name=.venv_hls4ml
````

Now to start a jupyter session type 
```
jupyter-lab
```

Otherwise just activate the virtual enviroment with
```
source venv_hls4ml/bin/activate
```
